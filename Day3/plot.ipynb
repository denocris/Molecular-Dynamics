```python
import numpy as np
from pylab import *
```

```python
def readfile(filename):
    ret = open("%s" %(filename),"r")
    ret = [line for line in ret]
    ret = [float(line.strip()) for line in ret]
    return ret

time_task_A = readfile("data/time_A.dat") 
time_task_B = readfile("data/time_B.dat") 

nprocs = [1,2,4,8,16,20]

plt.figure(figsize=(14,10))
plt.title("Scaling of the simplemd code")
plt.plot(nprocs,time_task_A, "--o", color = "red", label = "parallel compute force")
plt.plot(nprocs,time_task_B, "--o", color = "blue", label = "parallel compute force AND compute list")
plt.legend(loc="upper right", fontsize=15)
plt.text(14, 235, r'./lattice $16$ , $natoms =16384$', fontsize=15)
plt.savefig('scaling.png')
plt.show()
```
